```python
import pandas as pd

df = pd.read_csv(
    "data/taxationbydistrict.csv",
    sep=";",
    encoding="ISO-8859-1",
    skiprows=7,
)

df.columns = [
    "Year",
    "Region_Code",
    "Region_Name",
    "Taxpayer_Count",
    "Total_Income_KEuros",
    "Total_Taxes_KEuros",
]


print(df.head())
```
